# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Prepare the Data

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)
z = np.sin(x)
```

## 2. Create Plot

### 2.1 One Figure

```python
# create figure object
plt.figure()

# set x, y vector data
plt.plot(x, y)

# display figure
plt.show()
```

### 2.2 One Figure - Several lined subplots

```python
# create figure and axes objects plt.subplots(n rows, n columns)
fig, axs = plt.subplots(2)

# set data in subplots
axs[0].plot(x, y)
axs[1].plot(x, z)

# display figure
plt.show()
```

### 2.3 One Figure - A matrix of subplots

```python
# create figure and axes objects plt.subplots(n rows, n columns)
fig, axs = plt.subplots(2, 2)

# set data in subplots
axs[0, 0].plot(x, y)
axs[0, 1].plot(x, z)
axs[1, 0].plot(x, z)
axs[1, 1].plot(x, y)

# display figure
plt.show()
```

## 3. Choose Graphical Representation

### 3.1 Scatter

```python
plt.figure()
plt.scatter(x, y)
plt.show()
```

### 3.2 Histogram

```python
plt.figure()
hist = plt.hist(y)
```

### 3.3 Box-plot

```python
plt.figure()
box = plt.boxplot(y)
```

## 4. Customize Plot

### 4.1 Colors, Color Bars & Color Maps

```python
# create figure object, define figsize
plt.figure(figsize=(15,3))

# title
plt.title('Figure 1')

# indicate x, y vector data, colored line
plt.plot(x, y, c='b')

# axis limits
plt.xlim(0, 10)
plt.ylim(-1, 1)

# display plot
plt.show()
```

```python
# create figure object
fig, axs = plt.subplots(2, figsize=(10,7))

# title
axs[0].set_title('Figure 1 - Subplot 1')
axs[1].set_title('Figure 1 - Subplot 2')

# indicate x, y vector data, colored line, linewidth, linestyle
axs[0].plot(x, y, c='b', linewidth=4)
axs[1].plot(x, z, c='r', linestyle='--')

# axis limits
axs[0].set_xlim(0, 10)
axs[1].set_xlim(0, 10)

axs[0].set_ylim(-1, 1)
axs[1].set_ylim(-1, 1)

# display plot
plt.show()
```
