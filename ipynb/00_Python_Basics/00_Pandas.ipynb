# Pandas Data Structures

```python
import numpy as np
import pandas as pd
import datetime
```

## 1. Create DataFrame

```python
# temporal data series hourly spaced
x = np.arange(datetime.datetime(2000, 1,1), datetime.datetime(2000, 1,20), datetime.timedelta(hours=1))

# random numbers between 0-1 of lenght = len(x)
y = np.random.rand(len(x))
```

```python
data = pd.DataFrame(
    {
        'time': x,
        'col1': y,
        'col2': y**2
    }
)

# display first rows of pd DataFrame
data.head()
```

## 1.2  Setting/Resetting index

```python
data = data.set_index('time')
data.head()
```

## 1.3 Accesing to columns

```python
# get ['col1']
data[['col1']].head()
```

```python
# get column1 values
data['col1'].values
```

## 1.4 Selecting, Boolean Indexing & Setting

### by position - iloc

```python
data.iloc[[0]]
```

### by label - loc

```python
data.loc[['2000-01-01 17:00:00']]
```

```python
data.loc[data.col1>.7]
```

```python
data.loc[(data.col1>.7) & (data.col2>.8)]
```
